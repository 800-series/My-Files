```python
import pandas as pd
import h5py
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
filename = './test_catvnoncat.h5'
dictionary = {}
with h5py.File(filename, "r") as f:
    for key in f.keys():
        print(key)

        ds_arr = f[key][()]   # returns as a numpy array
        dictionary[key] = ds_arr # appends the array in the dict under the key
```

```python
dictionary['test_set_x'].shape
```

```python
dictionary['test_set_y'].shape
```

```python
dictionary['list_classes']
```

```python
index = 16
plt.imshow(dictionary['test_set_x'][index])
```

```python
print ("y = " + str(dictionary['test_set_y'][index]) + ", it's a '" + np.squeeze(dictionary['list_classes'][dictionary['test_set_y'][index]]).decode("utf-8") +  "' picture.")
```

```python
index = 15
plt.imshow(dictionary['test_set_x'][index])
print ("y = " + str(dictionary['test_set_y'][index]) + ", it's a '" + np.squeeze(dictionary['list_classes'][dictionary['test_set_y'][index]]).decode("utf-8") +  "' picture.")
```
